## Exploratory Data Analysis(Chapter 6)
EDA stands for Exploratory Data Analysis. It is a process of analyzing and summarizing a data set in order to gain insights and understanding about the data. The goal of EDA is to uncover the underlying structure and relationships in the data, and to identify any patterns, trends, or anomalies that may be present.

EDA is an iterative process that typically involves the following steps:

Importing the data: reading the data into a software program, such as a spreadsheet or data analysis software.

Cleaning the data: checking for missing values, correcting any errors, and transforming the data as necessary to ensure that it is in a suitable format for analysis.

Summarizing the data: computing basic statistics, such as mean, median, and standard deviation, and creating visualizations, such as histograms, box plots, and scatter plots, to help understand the distribution of the data.

Identifying relationships: looking for correlations, patterns, and trends in the data, and using visualizations to help identify any relationships that may be present.

Modeling the data: using statistical models to explain relationships in the data and make predictions.

The process of EDA is important because it allows you to understand the structure and properties of your data, and it can help you identify any limitations or biases in the data that may affect your analysis. Additionally, EDA can help you identify the most important variables in your data and identify the most appropriate methods for modeling and analyzing the data.

### Satistical data types:
There are two main types of data in statistics: numerical data and categorical data.

Numerical data: Numerical data is data that can be measured or quantified, such as height, weight, age, or temperature. There are two types of numerical data: continuous data and discrete data. Continuous data can take any value within a given range, while discrete data can only take specific values within a given range.

Categorical data: Categorical data is data that can be divided into categories or groups. Categorical data is often nominal or ordinal in nature, meaning that the categories do not have any inherent order or magnitude. Examples of categorical data include gender, eye color, and types of fruit.

In addition to these two main types of data, there is also a third type of data called ordinal data, which is a type of categorical data that has an inherent order. For example, a person's education level (high school, college, graduate school) is an example of ordinal data because there is a clear order to the categories.

It's important to understand the type of data you are working with, as different types of data require different types of analysis and modeling techniques. For example, you might use a different method to analyze numerical data than you would use to analyze categorical data

### Properties of data:
1. Granularity (Granularity refers to the level of detail in the data. Data can be fine-grained, meaning that it contains a lot of detailed information, or it can be coarse-grained, meaning that it contains only high-level information.)

2. Scope (Scope refers to the extent of the data. This can refer to the geographical, subject matter, or temporal range of the data. For example, data may be collected about a specific region, industry, or time period.)

3. Temporality (Temporality refers to the time aspect of the data. This can include the time frame in which the data was collected, as well as the frequency of data collection. For example, data may be collected in real-time, daily, weekly, monthly, or less frequently.)

4. Faithfulness (Faithfulness refers to the accuracy and reliability of the data. This includes factors such as measurement error, data quality, and the representativeness of the sample. Good faithfulness means that the data is a true and reliable representation of the phenomenon being measured.)

### What is Data Cleaning? (Chapter 7)

Data cleaning, also known as data cleansing or data wrangling, is the process of identifying and correcting or removing inaccuracies, inconsistencies, and missing or irrelevant information in a dataset. The goal of data cleaning is to prepare the data for analysis or modeling by making it as accurate, consistent, and complete as possible.

Data cleaning is a critical step in the data processing pipeline, as even small amounts of inaccuracies or inconsistencies in the data can lead to incorrect or biased results. Data cleaning tasks can include:

Removing duplicate records
Handling missing values
Handling outliers
Correcting inaccuracies or inconsistencies in the data
Removing irrelevant or redundant information
Transforming data into a consistent format
Combining multiple datasets into a single dataset

### Bootstrap a confidence interval

Bootstrapping is a widely used technique in data science for a variety of purposes. Here are some examples:

Estimating confidence intervals: Bootstrapping can be used to estimate the uncertainty around a point estimate of a population parameter (e.g., mean, median, correlation, etc.). This can be useful in hypothesis testing or when making predictions, as it helps to quantify the degree of uncertainty in the estimate.

Model selection and validation: Bootstrapping can be used to evaluate the performance of different models by estimating the variability of their predictions on different subsets of the data. This can help data scientists choose the best model for a given problem.

Outlier detection: Bootstrapping can be used to identify outliers in a dataset by comparing the distribution of a statistic (e.g., mean, standard deviation, etc.) calculated from the original dataset to the distribution of the same statistic calculated from bootstrapped datasets. Outliers are identified as values that fall outside the range of values observed in the bootstrapped distributions.

Data imputation: Bootstrapping can be used to impute missing values in a dataset by creating multiple imputed datasets and combining them to obtain a single dataset that reflects the variability of the imputed values.

Overall, bootstrapping is a powerful and flexible technique that can be used in many different applications in data science.


### Chapter 10 Classification
Classification is a type of supervised learning in data science that involves predicting which category, or class, a new observation belongs to based on its characteristics. In other words, it is the process of identifying to which category a particular data point belongs.

The goal of classification is to build a model that can accurately classify new observations based on patterns found in the training data. The training data is a set of labeled examples, where each example has a set of features and a corresponding label or category.

The classification algorithm learns from the training data to recognize patterns and relationships between the features and the labels. Once the model is trained, it can be used to predict the label of new, unseen data points.

Examples of classification problems include spam detection in email, sentiment analysis of social media posts, and image recognition. Popular classification algorithms include decision trees, logistic regression, support vector machines, and neural networks.

### The k-nearest neighbor algorithm (k-NN)
The k-nearest neighbor algorithm (k-NN) is a supervised learning algorithm in machine learning used for classification and regression problems. It is a simple and powerful algorithm that can be used for both binary and multiclass classification.

In k-NN, the "k" represents the number of nearest neighbors to consider. The algorithm works by calculating the distance between the new observation and all the observations in the training set. It then selects the k closest points to the new observation and assigns the class of the majority of those points to the new observation.

The distance metric used in k-NN can vary depending on the data and problem. Common distance metrics include Euclidean distance, Manhattan distance, and Minkowski distance.

One of the advantages of the k-NN algorithm is that it can work well with small datasets and is relatively easy to implement. However, one of the disadvantages of k-NN is that it can be computationally expensive and may require a large amount of memory when dealing with large datasets.

Overall, the k-NN algorithm is a useful tool in the data scientist's toolbox for solving classification problems.
